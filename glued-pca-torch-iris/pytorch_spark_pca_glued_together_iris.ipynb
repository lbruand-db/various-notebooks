# Glueing a dimension reduction upstream a torch model

We perform dimensionality reduction using Principal Component Analysis (PCA) to project our high-dimensional data into a lower-dimensional space. After reducing the dimensions, we train a PyTorch model on the dimension-reduced vectors.

We then show that we can rebuild a end-to-end torch model from the original original features without any retraining.

We will be using the IRIS dataset as its low dimensionality makes calculation quick. But of course dimension reduction is only making sense for higher dimensions than this. (IRIS n=4).
We have chosen a PCA but any kind of dimensionality reduction technic could apply here as long as you can turn them into a torch model.

Tested on DBR ML 15.4

````python
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target
````

````python
import torch
X_tensor = torch.tensor(X, dtype=torch.float32)
y_tensor = torch.tensor(y, dtype=torch.long)
print('Features sample:', X_tensor[:5])
print('Target sample:', y_tensor[:5])
````

````python
from sklearn.model_selection import train_test_split
# Split the original data and labels before PCA to avoid data leakage
X_train, X_test, y_train, y_test = train_test_split(X_tensor, y_tensor, test_size=0.2, random_state=42)
print(f"Train shape: {X_train.shape}, Test shape: {X_test.shape}")
````

# Fit the PCA

````python
import pandas as pd
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.feature import PCA, StandardScaler
import numpy as np

# Use only training data for PCA fitting
X_train_np = X_train.numpy()
df_pd_train = pd.DataFrame(X_train_np)
df_spark_train = spark.createDataFrame(df_pd_train)

from pyspark.ml import Pipeline
from pyspark.ml.feature import StandardScaler

assembler = VectorAssembler(
    inputCols=df_spark_train.columns,
    outputCol="features"
)
scaler = StandardScaler(
    inputCol="features",
    outputCol="scaledFeatures",
    withMean=True,
    withStd=True
)
pca = PCA(
    k=3,
    inputCol="scaledFeatures",
    outputCol="pcaFeatures"
)
pipeline = Pipeline(stages=[assembler, scaler, pca])
pca_model = pipeline.fit(df_spark_train)

# Transform both train and test sets
result_train = pca_model.transform(df_spark_train)
X_pca_train_tensor = torch.tensor(
    np.array([row['pcaFeatures'] for row in result_train.select('pcaFeatures').collect()]), dtype=torch.float32)

X_test_np = X_test.numpy()
df_pd_test = pd.DataFrame(X_test_np)
df_spark_test = spark.createDataFrame(df_pd_test)
result_test = pca_model.transform(df_spark_test)
X_pca_test_tensor = torch.tensor(
    np.array([row['pcaFeatures'] for row in result_test.select('pcaFeatures').collect()]), dtype=torch.float32)

explained_variance = pca_model.stages[2].explainedVariance.toArray()
````

````python
print('Explained variance ratio:', explained_variance)
print('PCA transformed sample:', X_pca_train_tensor[:5])
````

````python
# Convert Spark PCA model's principal components (model.pc) to a torch tensor
import torch

scaler_std = torch.tensor(pca_model.stages[1].std, dtype=torch.float32)
scaler_mean = torch.tensor(pca_model.stages[1].mean, dtype=torch.float32)
# model.pc is a Spark DenseMatrix; convert to a NumPy array first
pca_components = torch.tensor(pca_model.stages[2].pc.toArray(), dtype=torch.float32)

print(scaler_std)
print(scaler_mean)
print(pca_components)
````

# Train a pytorch model on top of the PCA

````python
import torch.nn as nn
import torch.nn.functional as F

class SimpleClassifier(nn.Module):
    """
    SimpleClassifier is a toy neural network model for classifying iris flowers.
    Args:
        None
    """
    def __init__(self):
        super(SimpleClassifier, self).__init__()
        self.fc1 = nn.Linear(X_pca_train_tensor.shape[1], 16)
        self.fc2 = nn.Linear(16, 3)  # 3 classes in IRIS
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x

model = SimpleClassifier()
````

````python
import torch.optim as optim


criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

n_epochs = 100
for epoch in range(n_epochs):
    model.train()
    optimizer.zero_grad()
    outputs = model(X_pca_train_tensor)
    loss = criterion(outputs, y_train)
    loss.backward()
    optimizer.step()
    if (epoch+1) % 20 == 0:
        print(f"Epoch {epoch+1}/{n_epochs}, Loss: {loss.item():.4f}")

# Evaluate
model.eval()
with torch.no_grad():
    test_outputs = model(X_pca_test_tensor)
    _, predicted = torch.max(test_outputs, 1)
    accuracy = (predicted == y_test).float().mean().item()
    print(f"Test Accuracy: {accuracy:.2%}")
````

# Build a single torch model that comprises the PCA (as a torch layer) and the original classifier

````python
import torch
import torch.nn as nn

class TorchPCA(nn.Module):
    """
    TorchPCA applies a linear PCA transformation using parameters from a fitted sklearn PCA.
    Args:
        components (torch.Tensor): PCA components (eigenvectors), shape [n_features, n_components].
        mean (torch.Tensor): Mean vector used for centering and scaling, shape [n_features].
        std (torch.Tensor): Standard deviation vector used for centering and scaling, shape [n_features].
    Forward:
        x (torch.Tensor): Input features, shape [batch_size, n_features].
        Returns: Transformed features in PCA space, shape [batch_size, n_components].    
    """
    def __init__(self, components, mean, std):
        super().__init__()
        self.register_buffer('components', components)
        self.register_buffer('mean', mean)
        self.register_buffer('std', std)
    def forward(self, x):
        x_centered = x - self.mean
        x_scaled = x_centered / self.std
        return torch.matmul(x_scaled, self.components)

# Glue TorchPCA and SimpleClassifier
class PCAGluedClassifier(nn.Module):
    """
    PCAGluedClassifier applies PCA transformation followed by classification using a neural network. The resulting model is torch module that could be retrained if need be.
    Args:
        pca_module (TorchPCA): Module for PCA transformation.
        classifier (SimpleClassifier): Neural network classifier.
    Forward:
        x (torch.Tensor): Input features, shape [batch_size, n_features].
        Returns: Class probabilities, shape [batch_size, n_classes].
    """
    def __init__(self, pca_module, classifier):
        super().__init__()
        self.pca = pca_module
        self.classifier = classifier
    def forward(self, x):
        x_pca = self.pca(x)
        return self.classifier(x_pca)

# Instantiate the glued model
pca_torch = TorchPCA(pca_components, scaler_mean, scaler_std)
glued_model = PCAGluedClassifier(pca_torch, model)

# Example usage: glued_model(X_tensor)
````

# Parallel SHAP on whole model using pandas UDF

````python
import shap
import torch.nn.functional as F
from pyspark.sql.functions import pandas_udf, col
from pyspark.sql.types import ArrayType, DoubleType, StructType, StructField
import pandas as pd
import numpy as np

# Create Spark DataFrame from the sample data
X_sample = X_tensor  # Use more samples since we're parallelizing
X_sample_np = X_sample.numpy()

# Convert to Spark DataFrame with feature columns
feature_cols = [f"feature_{i}" for i in range(X_sample_np.shape[1])]
X_df = spark.createDataFrame(
    pd.DataFrame(X_sample_np, columns=feature_cols)
)

# Define the schema for SHAP values output
shap_schema = ArrayType(DoubleType())

# Define pandas UDF for SHAP computation
@pandas_udf(returnType=shap_schema)
def compute_shap_values(feature_batch: pd.Series) -> pd.Series:
    """
    Compute SHAP values for a batch of features using pandas UDF.
    This runs in parallel across Spark workers.
    """
    # Convert Series of arrays back to numpy array
    batch_array = np.vstack(feature_batch.values)
    
    # Predict function for SHAP
    def model_predict_batch(x_numpy):
        x_tensor = torch.tensor(x_numpy, dtype=torch.float32)
        with torch.no_grad():
            logits = glued_model(x_tensor)
            probs = F.softmax(logits, dim=1)
            return probs.numpy()
    
    # Create SHAP explainer with a small background sample
    background_sample = batch_array[:10]  # Use first 10 samples as background
    explainer = shap.Explainer(model_predict_batch, background_sample)
    
    # Compute SHAP values for the batch (focusing on class 0)
    shap_values = explainer(batch_array, max_evals=50)
    
    # Return SHAP values for class 0 as a list for each row
    return pd.Series([shap_val.tolist() for shap_val in shap_values.values[:,:,0]])

# Combine all features into a single array column for the UDF
from pyspark.sql.functions import array
X_df_with_array = X_df.withColumn(
    "features_array", 
    array(*[col(c) for c in feature_cols])
)

# Apply the SHAP UDF in parallel
shap_results = X_df_with_array.select(
    compute_shap_values(col("features_array")).alias("shap_values")
).collect()

# Extract SHAP values and convert back to numpy for plotting
shap_values_array = np.array([row.shap_values for row in shap_results])

print(f"Computed SHAP values for {len(shap_values_array)} samples using parallelized pandas UDF")
print(f"SHAP values shape: {shap_values_array.shape}")

# Create summary plot
shap.summary_plot(shap_values_array, X_sample_np, feature_names=iris.feature_names)
````
