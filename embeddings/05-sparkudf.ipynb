Note : tested on Single Node DBR 16.4 LTS ML Cluster with g5.xlarge (A10G). Not working on serverless GPU

```python
import mlflow

catalog_name = "databricks_e5_v2_models"
model_name = "e5_large_v2"
embed = mlflow.pyfunc.spark_udf(spark, f"models:/{catalog_name}.models.{model_name}/1", "array<float>")
```

```python
# Create a single row Spark DataFrame with the specified text
df = spark.createDataFrame([("What's Apache Spark",)], ["text"])

# Display the DataFrame
display(df)
```

```python
embedded_df = df.select(embed(df.text).alias('embedding'))
```
